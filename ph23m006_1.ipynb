Problem 2.1

a. Let H ∈ {1, . . . , K} be a discrete random variable, and let e1 and e2 be the observed values of two other
random variables E1 and E2.

Suppose we wish to calculate the vector

$\overrightarrow{P}(H \vert e1,e2) =  (P(H=1 \vert e1,e2),.....,P(H=K \vert e1,e2))
$

Which of the following sets of numbers are sufficient for the calculation?

i. P(e1, e2), P(H), P(e1|H), P(e2|H)

ii. P(e1, e2), P(H), P(e1, e2|H)

iii. P(e1|H), P(e2|H), P(H)

b. Now suppose we now assume E1 ⊥ E2|H (i.e., E1 and E2 are conditionally independent given H). Which
of the above 3 sets are sufficient now?

\textbf{Ans.}

The vector given to us is-

$\overrightarrow{P} =  P(H=1 \vert e1,e2),.....,P(H=K \vert e1,e2)
$

Bayes theorem allows us to calculate


$P(H=K \vert e1,e2)= \frac {(P(H=K)P(e1,e2 \vert H))}{P(e1,e2)}
$

so option ii is sufficient to solve question a.

For question b, the following relation can be written

$P(H \vert e1,e2) = \frac {P(H)P(e1 \vert H) P(e2 \vert H)}{P(e1,e2)}$

thus making (i) and (ii) sufficient to solve it. However,

Since

\begin{align*}
P(e1,e2) = \sum_{H}P(e1,e2,H)
\\ = \sum_{H}P(H)P(e1 \vert H)P(e2 \vert H)
  \end{align*}
(iii) also becomes sufficient as we can calculate P(e1,e2) independently.

 Problem 2.5

 Suppose X, Y are two points sampled independently and uniformly at random from the interval [0, 1]. What is the expected location of the leftmost point?

 Ans.

Let m denote the location of the leftmost point,

this gives us -
\begin{align*}
\\ p(m>x)=p([X>x]and[Y.x])
\\ = p(X>x)p(Y>x)
\\ = (1-x)^2
\end{align*}

Now we can write-
\begin{align*}
\\ \mathbb E[m] = \int x.p(m=x)dx
\\ = \int p(m>x)dx
\\ = \int_0^1 (1-x)^2dx
\\ = \frac{1}{3}
\end{align*}
